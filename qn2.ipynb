## Graphical Comparison of Household Incomes

### Introduction:
This notebook presents a graphical comparison of household incomes between members and non-members of farmer organizations using the provided dataset. The purpose is to visually assess the distribution of incomes within these two groups.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the corrected data
data = pd.read_csv("qn1_corrected_data_exercise.csv")
```

### Graphical Representation:
A boxplot was created to compare the distribution of household incomes:

- **Members of Farmer Organizations:** The box represents the interquartile range (IQR) of income, with the line inside the box indicating the median income.

- **Non-Members of Farmer Organizations:** Similarly, the box represents the IQR, and the line inside the box represents the median income.


```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("qn1_corrected_data_exercise.csv")

members_incomes = data[data['fo'] == 1]['income'].values
non_members_incomes = data[data['fo'] == 0]['income'].values
```

## Figure 1: Box Plot

### Title: Distribution of Household Incomes (Box Plot)

**Interpretation (Figure 1):**
- The box plot visually compares the distribution of household incomes between members and non-members of farmer organizations.
- The blue box represents the interquartile range (IQR) for non-members, indicating where the middle 50% of incomes fall. The line inside the box represents the median income for non-members.
- Similarly, the orange box represents the IQR for members, with the line inside indicating the median income for members.
- Outliers, if present, are shown as individual data points beyond the "whiskers." They represent incomes significantly different from the median.


```python
# Figure 1: Box Plot
plt.figure(figsize=(8, 6))
plt.boxplot([non_members_incomes, members_incomes], labels=["Non-Members", "Members"])
plt.title("Distribution of Household Incomes (Box Plot)")
plt.xlabel("Membership Status")
plt.ylabel("Income")
plt.show()
```

## Figure 2: Histogram

### Title: Distribution of Household Incomes (Histogram)

**Interpretation (Figure 2):**
- The histogram compares the distribution of household incomes for members (in orange) and non-members (in blue).
- The x-axis represents income values, while the y-axis shows the frequency or count of households in each income range.
- By comparing the two histograms, you can observe how incomes are distributed within each group. For example, you can see which group has more households with higher or lower incomes.


```python
# Figure 2: Histogram
plt.figure(figsize=(8, 6))
plt.hist([non_members_incomes, members_incomes], bins=20, alpha=0.7, label=["Non-Members", "Members"])
plt.title("Distribution of Household Incomes (Histogram)")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

## Figure 3: Density Plot

### Title: Distribution of Household Incomes (Density Plot)

**Interpretation (Figure 3):**
- The density plot shows the probability density of income values for members (in orange) and non-members (in blue).
- It provides a smoothed representation of the distribution, highlighting where the data is concentrated.
- By comparing the two density plots, you can observe the shape and concentration of incomes within each group. It allows you to see which group has a higher density of households at different income levels.


```python
# Figure 3: Density Plot
plt.figure(figsize=(8, 6))
plt.hist(non_members_incomes, bins=20, density=True, alpha=0.5, label="Non-Members")
plt.hist(members_incomes, bins=20, density=True, alpha=0.5, label="Members")
plt.title("Distribution of Household Incomes (Density Plot)")
plt.xlabel("Income")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

## Figure 4: Violin Plot

### Title: Distribution of Household Incomes (Violin Plot)

**Interpretation (Figure 4):**
- The violin plot combines aspects of a box plot and a kernel density estimate.
- It provides a more detailed view of the distribution of incomes within each group.
- The width of the violin at different points represents the density of data points at those income values.
- The white dot inside each violin represents the median income for that group.
- Violin plots allow you to compare the shape and spread of income distributions between members and non-members. You can identify where the data is most concentrated and the overall distribution characteristics for each group.

```python
# Figure 4: Violin Plot
plt.figure(figsize=(8, 6))
plt.violinplot([non_members_incomes, members_incomes], showmedians=True)
plt.xticks([1, 2], ["Non-Members", "Members"])
plt.title("Distribution of Household Incomes (Violin Plot)")
plt.xlabel("Membership Status")
plt.ylabel("Income")
plt.show()
```
